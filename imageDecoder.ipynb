```python
import matplotlib as plt
from PIL import Image


img = Image.open("birds.jpg")

image = img.convert("RGB")


def imageResizeData(sizeX, sizeY, image):
    rgbData = []
    newImg = image.resize((sizeX, sizeY), Image.LANCZOS)
    for i in range(sizeX):
        for j in range(sizeY):
            rgbData.append(newImg.getpixel((j,i)))
    return rgbData

def unpack(RGB):
    red = []
    green = []
    blue = []
    for i in RGB:
        red.append(i[0])
        green.append(i[1])
        blue.append(i[2])
    return red, green, blue


def imageResize(sizeX, sizeY, image):
    newImg = image.resize((sizeX, sizeY), Image.LANCZOS)
    return newImg

data = imageResizeData(128, 128, image)
newImage = imageResize(128, 128, image)

```
